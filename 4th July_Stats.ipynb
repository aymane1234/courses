```python
import numpy as np
import statsmodels.api as sm 
import pylab as py
```

```python
data = np.random.normal(0,1,100) #standard  normal distribution
```

```python
sm.qqplot(data, line= '45') 
py.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import powerlognorm
numargs = powerlognorm.numargs
a,b = 4.32, 3.18
rv = powerlognorm(a,b)
print('RV: \n',rv)

quantile  = np.arange(0.01,1,0.1)

#Random variabe
R = powerlognorm.rvs(a,b)
#PDF
pdf = powerlognorm.pdf(a,b,quantile)
```

```python
from scipy import stats
```

```python
#Generate non-normal data "exponential"
original_data = np.random.exponential(size=1000)
```

```python
#Apply a boxcox transformation
fitted_data, fitted_lambda = stats.boxcox(original_data)
```

```python
#plot the data
fig, ax = plt.subplots(1,2)
sns.displot(original_data, kde= True)
sns.displot(fitted_data,kde = True)
```

```python
#See whether it fits the straight line by applying QQ-PLot
sm.qqplot(fitted_data,line = '45')
py.show()
```
