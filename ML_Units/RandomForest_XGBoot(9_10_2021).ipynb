```python
import pandas as pd 
import numpy as np
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
x = df.drop(columns='quality')
y = df.quality
```

```python
y
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state= 0)
```

```python
bag_dt = BaggingClassifier(DecisionTreeClassifier(),n_estimators=100) #inside one bag there will be multiple Dtrees 10  "models in general and get the best to do the prediction by voting"
```

```python
bag_dt.fit(x_train,y_train)
```

```python
bag_dt.predict(x_test)
```

```python
bag_dt.base_estimator_
```

```python
bag_dt.classes_
```

```python
bag_dt.estimator_params
```

lets give it some params


```python
bag_dt = BaggingClassifier(DecisionTreeClassifier(criterion='gini'),n_estimators=100)
```

```python
bag_dt.estimator_params
```

Lets try to create a different bag with a differten model namely KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
bag_knn  =BaggingClassifier(KNeighborsClassifier(6), n_estimators=100 )  #Use instead of Dtrees KNN
```

```python
bag_knn.fit(x_train,y_train)
```

```python
bag_knn.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=5)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
bag_dt.score(x_test,y_test)
```

```python
bag_knn.score(x_test,y_test)
```

```python
plt.figure(figsize=(20,20))
for i in range(len(rf.estimators_)):
  tree.plot_tree(rf.estimators_[i],filled=True)
```

```python
grid_param = {"n_estimators" : [5,20,50,100,120,150],
              "criterion" : ['gini','entropy'],
              "max_depth" : range(10),
              "min_samples_leaf": range(10),
                  
}
```

```python
grid_rf = GridSearchCV(param_grid = grid_param, cv=5,n_jobs=6, verbose=1,estimator=rf)
```

```python
grid_rf.fit(x_train,y_train)
```

```python
grid_rf.best_params_
```
