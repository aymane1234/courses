```python
df = pd.read_csv('winequality-red.csv')
```

```python
x = df.drop(columns='quality')
y = df.quality
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state= 0)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
param = {'n_neighbours': [3,5,9,13,16,21,25],
         'p': [1,2],
         'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
         'leaf_size': [10,15,20,25,30],
         'weight': ['uniform', 'distance']
         }
```

```python
grid_cv = GridSearchCV(estimator = knn, param_grid= param)
```

```python
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.get_params
```

```python
knn_opt =KNeighborsClassifier(algorithm='auto', leaf_size=10, metric='minkowski',metric_params=None, n_jobs=None, n_neighbors=21, p=1,weights='uniform')
                     
```

```python
knn_opt.fit(x_train,y_train)
```

```python
knn_opt.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
svc.score(x_test,y_test)
```

```python
param_s = {'kernel': [ 'linear', 'poly', 'rbf', 'sigmoid'],
           'C' : [.1,.4,.6,1,2,3,100,200,500],
           'gamma':[.001,.1,.4,.004,.003]
}
svm_grid = GridSearchCV(svc,param_grid=param_s,verbose=3)
svm_grid.fit(x_train,y_train)
```

```python
svm_grid.best_params_
```

```python
svc1 =SVC(kernel='linear')
svc1.fit(x_train,y_train)
```
