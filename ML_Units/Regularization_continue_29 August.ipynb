```python
#adjusted r square
def adj_r2(x,y):
  r2 = lr.score(x,y)
  n = x.shape[0]
  p = x.shape[1]
  adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
  return adjusted_r2
```
